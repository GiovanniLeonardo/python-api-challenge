```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from config import g_key

gmaps.configure(api_key=g_key)
```

```python
# Store Part I results into DataFrame
csv_file = "../WeatherPy/output_data/cities.csv"
weather_df = pd.read_csv(csv_file)
weather_df.head()
# Load the csv exported in Part I to a DataFrame
```

### Humidify Heatmap

```python
# location variable to store latitude and longitude
location = weather_df[["Latitude", "Longitude"]]

# humidity variable to store Humidity
humidity = weather_df["Humidity"].astype(float)
```

```python
# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(location, weights=humidity, max_intensity= 100, point_radius = 3)

# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
##NOTE: Display Map would not work, i get msg API not activated but I generated a new key and I added the jupyter nbextension enable --py gmaps
```

```python
### Create new DataFrame fitting weather criteria
<li> Narrow down the cities to fit weather conditions.</li>
<li> Drop any rows will null values.</li>
```

```python
great_weather_df = weather_df.loc[(weather_df['Temperature_Max']  >= 70)
               & (weather_df['Temperature_Max']  < 85)
               & (weather_df['Cloudiness']  < 40)
               & (weather_df['Humidity'] < 50)
                          ]
great_weather_df = great_weather_df.dropna()
great_weather_df = great_weather_df.reset_index()
great_weather_df
```

```python
### Hotel Map
<li> Store into variable named hotel_df.</li>
<li> Add a "Hotel Name" column to the DataFrame.</li>
```

```python
hotel_df = great_weather_df
hotel_df['Hotel_Name'] = ""
hotel_df
```

```python
hotels = []

for i in range(len(hotel_df)):
    Latitude = hotel_df.loc[i]['Latitude']
    Longitude = hotel_df.loc[i]['Longitude']

    params = {
        "location": f"{Latitude},{Longitude}",
        "types" : "hotel",
        "radius": 5000,
        "key": g_key
    }
    
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    requested = requests.get(base_url, params=params)
    jsn = requested.json()
    try:
        hotels.append(jsn['results'][0]['name'])
    except:
        hotes.append("")
hotel_df["Hotel_Name"] = hotels
hotel_df = hotel_df.dropna()
hotel_df.head()
```

```python
### Using the template add the hotel marks to the heatmap
```

```python
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel_Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""

hotel = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
location = hotel_df[["Latitude", "Longitude"]]
```

```python
# mark the layer ontop of heat map
markers = gmaps.marker_layer(location, info_box_content=hotel)
fig.add_layer(markers)

# Display Map would not work, i get msg API not activated but I generated a new key and I added the jupyter nbextension enable --py gmaps
fig
```
